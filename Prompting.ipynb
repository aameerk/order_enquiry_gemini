```python
import PIL.Image
img = PIL.Image.open('/content/tax-invoice-format-2-tallyprime.jpg')
img
```

```python
prompt = """ This image contains details on tax invoice with product price etc. Given the image, describe it as thoroughly
as possible based on what you see in the image, making sure to note all of the product features. Return output in json format:{description: description, features: {feature1, feature2, feature3, etc}} """
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')
response = model.generate_content([prompt, img])
print(response.text)
```
